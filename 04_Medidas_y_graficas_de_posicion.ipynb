```python
# Calcula los cuartiles utilizando la función percentile de NumPy
q1 = np.percentile(datos, 25)
q2 = np.percentile(datos, 50)
q3 = np.percentile(datos, 75)

plt.axvline(q1, color='red', linestyle='dashed', label='Q1')
plt.axvline(q2, color='green', linestyle='dashed', label='Q2 (Mediana)')
plt.axvline(q3, color='blue', linestyle='dashed', label='Q3')

# Crea un histograma para visualizar la frecuencia de los datos
plt.hist(datos, bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencia de Datos')
plt.grid(True)
plt.show()
```

## **Deciles**

```python
def deciles_desde_cero(datos):
  '''
  datos ---> list : información
  '''
  # Longitud de lista
  n = len(datos)
  # Organizar la información
  datos.sort()

  if n % 2 == 0:
    deciles = []
    for k in range(1,10):
      deciles.append( datos[int(((k * n) / 10) - 1)] )
  else:
    deciles = []
    for k in range(1,10):
      deciles.append( datos[int(((k * (n+1)) / 10) - 1)] )

  return deciles
```

```python
deciles = deciles_desde_cero(datos)
deciles
```

```python
for i in range(0,9):
  decil = deciles[i]
  plt.axvline(decil, color='red', linestyle='dashed', label=str(i)+'%')

# Crea un histograma para visualizar la frecuencia de los datos
plt.hist(datos, bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencia de Datos')
plt.grid(True)
plt.show()
```

```python
# Imprime deciles
print(f'Los deciles son: {deciles}')

for i in range(0,9):
  decil = deciles[i]
  plt.axvline(decil, color='red', linestyle='dashed', label=str((i+1)*10)+'%')

# Crea un gráfico de caja con los deciles
plt.boxplot(datos, vert=False)
plt.scatter(deciles, np.ones_like(deciles), color='red', marker='o', label='Deciles')
plt.xlabel('Datos')
plt.ylabel('Deciles')
plt.title('Deciles de un Conjunto de Datos')
plt.legend()
plt.show()
```

## **Percentiles**

```python
def percentiles_desde_cero(datos):
  '''
  datos ---> list : información
  '''
  # Longitud de lista
  n = len(datos)
  # Organizar la información
  datos.sort()


  percentiles = []

  if n % 2 == 0:
    for k in range(1,100):
      percentiles.append( datos[int(((k * n) / 100) - 1)] )
  else:
    for k in range(1,100):
      percentiles.append( datos[int(((k * (n+1)) / 100) - 1)] )

  return percentiles
```
