```python
def deciles_desde_cero(datos):
  '''
  datos ---> list : información
  '''
  # Longitud de lista
  n = len(datos)
  # Organizar la información
  datos.sort()

  if n % 2 == 0:
    deciles = []
    for k in range(1,10):
      deciles.append( datos[int(((k * n) / 10) - 1)] )
  else:
    deciles = []
    for k in range(1,10):
      deciles.append( datos[int(((k * (n+1)) / 10) - 1)] )

  return deciles
```

## **Percentiles**

```python
def percentiles_desde_cero(datos):
  '''
  datos ---> list : información
  '''
  # Longitud de lista
  n = len(datos)
  # Organizar la información
  datos.sort()


  percentiles = []

  if n % 2 == 0:
    for k in range(1,100):
      percentiles.append( datos[int(((k * n) / 100) - 1)] )
  else:
    for k in range(1,100):
      percentiles.append( datos[int(((k * (n+1)) / 100) - 1)] )

  return percentiles
```

### **Los cuartiles**

Los cuartiles son valores estadísticos que dividen un conjunto de datos ordenados en cuatro partes iguales de acuerdo a la cantidad de datos. Hay tres cuartiles principales: el primer cuartil (Q1) que separa el 25% inferior de los datos del 75% superior, el segundo cuartil (Q2) que es igual a la mediana y divide los datos en dos partes iguales, y el tercer cuartil (Q3) que divide el 75% inferior del 25% superior de los datos. Los cuartiles son útiles para comprender la distribución y dispersión de un conjunto de datos, especialmente cuando se desea analizar la variabilidad dentro de ellos.

De este modo, si tenemos una colección de elementos $\{x_i\mid 1\leq i \leq n\}.$ Luego queremos determinar una formula general para hallar los cuartiles $Q_1,Q_2$ y $Q_3$. De este modo consideramos la paridad de la cantidad de datos para el estudio:

* $\mathbf{Q_1}:$
  * Si $\frac{n}{4}$ es entero, determinamos el primer cuartil como la media de los números en las posiciones $\frac{n}{4}$ y $\frac{n}{4} + 1.$ Es decir el valor será
  $$Q_1 = \frac{x_{\frac{n}{4}} + x_{\frac{n}{4} + 1}}{2}.$$
  * Si $\frac{n}{4}$ no es entero, redondeamos el número el número por arriba, así consideramos el indice $\lceil \frac{n}{4}\rceil.$ Así
  $$Q_1 = x_{\lceil \frac{n}{4}\rceil}.$$

* $\mathbf{Q_2}:$
  * Si $\frac{n}{2}$ es entero, determinamos el segundo cuartil como la media de los números en las posiciones $\frac{n}{2}$ y $\frac{n}{2} + 1.$ Es decir el valor será
  $$Q_2 = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2}.$$
  * Si $\frac{n}{2}$ no es entero, redondeamos el número el número por arriba, así consideramos el indice $\lceil \frac{n}{2}\rceil.$ Así
  $$Q_2 = x_{\lceil \frac{n}{2}\rceil}.$$

* $\mathbf{Q_3}:$
  * Si $\frac{3n}{4}$ es entero, determinamos el primer cuartil como la media de los números en las posiciones $\frac{3n}{4}$ y $\frac{3n}{4} + 1.$ Es decir el valor será
  $$Q_1 = \frac{x_{\frac{3n}{4}} + x_{\frac{3n}{4} + 1}}{2}.$$
  * Si $\frac{n}{4}$ no es entero, redondeamos el número el número por arriba, así consideramos el indice $\lceil \frac{3n}{4}\rceil.$ Así
  $$Q_1 = x_{\lceil \frac{3n}{4}\rceil}.$$

En el libro guía del curso podemos encontrar otra forma de calcular los cuartiles, por lo que se dice, que no hay un acuerdo universal sobre la mejor interpretación para calcular los cuartiles.
