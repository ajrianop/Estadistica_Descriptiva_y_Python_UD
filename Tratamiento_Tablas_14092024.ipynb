```python
goles = [52, 48, 50, 53, 47, 55, 49, 51, 54, 46,
         50, 52, 48, 53, 56, 49, 51, 54, 47, 50,
         52, 49, 55, 46, 51, 53, 48, 52, 50]
```

Vamos a calcular la media aritmetica para la cantidad de goles que anotaron 29 equipos en la temporada.

```python
media_aritmetica = sum(goles) / len(goles)
media_aritmetica
```

```python
# Valor mínimo
min(goles)
```

```python
_dict = {}
for numero_equipo in range(0, len(goles)):
  _dict[numero_equipo+1] = goles[numero_equipo]

_dict
```

```python
import pandas as pd
diccionario_goles = {'numero equipo': list(_dict.keys()), 'goles' : list(_dict.values())}
pd.DataFrame(diccionario_goles)
```

```python
diccionario_goles = {'numero': list(_dict.keys()), 'x_i' : list(_dict.values())}
df = pd.DataFrame(diccionario_goles)
```

```python
df
```

```python
df['x_i-media'] = df['x_i']-media_aritmetica
df
```

```python
df['|x_i-media|'] = abs(df['x_i-media'])
df
```

```python
desviacion_media = sum(df['|x_i-media|']) / len(df['numero'])
desviacion_media
```

```python
df['(x_i-media)**2'] = df['x_i-media']**2
df
```

```python
varianza = sum(df['(x_i-media)**2'])/len(df['numero'])
varianza
```

```python
desviacion_estandar = varianza ** (1/2)
desviacion_estandar
```

```python
coeficiente_variacion = (desviacion_estandar / media_aritmetica) * 100
print(f'El coeficiente de variación del problema es: {round(coeficiente_variacion,2)} %')
```

Vamos a determinar utilizando una función, si la variación es:
1. Baja : [0% , 30%],
2. Media : (30%, 70%],
3. Alta : Mayor al 70%.

```python
coeficiente_variacion = 46.1
print(f'El coeficiente de variacion es {coeficiente_variacion} lo que indica que,')
if abs(coeficiente_variacion) >= 0 and abs(coeficiente_variacion) <= 30:
  print('El coeficiente de variacion es BAJO.')
elif abs(coeficiente_variacion) > 30 and abs(coeficiente_variacion) <= 70:
  print('El coeficiente de variacion es MEDIO.')
else:
  print('El coeficiente de variacion es ALTA.')
```
