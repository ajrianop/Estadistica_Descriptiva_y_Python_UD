```python
import numpy as np

# Define mean and standard deviation for the normal distribution
mean = 6.5  # Mean of the distribution (midpoint of the range)
std_dev = 1.5  # Standard deviation of the distribution

# Generate 30 random numbers following a normal distribution
numbers_float = np.random.normal(mean, std_dev, 30)

# Clip the numbers to ensure they fall within the range of 3 to 10
numbers_float = np.clip(numbers_float, 3, 10)

# Round the floating-point numbers to the nearest integer
numbers_int = np.round(numbers_float).astype(int)

# Print the generated numbers
print(numbers_int)
```
