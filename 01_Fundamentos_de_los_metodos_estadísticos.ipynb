## **Calidad de los datos**

Hay que tener en cuenta que los métodos estadísticos no verifican, corrigen la deficiencia del dato, o mejoran el dato. De este modo tenemos que ser nosotros quienes verifiquen la calidad de los diferentes datos que se plantean en un estudio o investigación.

¿Qué componentes estudiar para observar la calidad de un conjunto de datos?

Los componentes principales a tener en cuenta serían:

* Representatividad.
* Confiabilidad.
* Validez.
* ...

**Representatividad**

Se refiere a la calidad de una muestra para reflejar adecuadamente las características y propiedades de la población a la que se refiere. En otras palabras, una muestra se considera representativa cuando los datos recopilados de esa muestra son un reflejo preciso de lo que existe en la población más amplia de la que se extrajo.

Cuando se trabaja con datos, es importante que la muestra sea representativa para poder hacer inferencias precisas y generalizables sobre la población completa. Si una muestra no es representativa, las conclusiones basadas en ella pueden ser sesgadas o inexactas.

La representatividad responde a la pregunta:
*¿Los resultados de la muestra pueden aplicarse o generalizarse a la población objeto de estudio?*

**Confiabilidad**

Se refiere a la consistencia y estabilidad de las mediciones, observaciones o respuestas en un conjunto de datos o instrumento de medición. En otras palabras, se refiere a la capacidad de obtener resultados consistentes y reproducibles cuando se realizan mediciones repetidas bajo condiciones similares. Una medida o conjunto de datos se considera confiable cuando produce resultados coherentes y predecibles a lo largo del tiempo y en diferentes situaciones.

La confiabilidad es esencial en la investigación científica y en la recopilación de datos, ya que garantiza que las mediciones sean precisas y que los resultados puedan ser confiables para tomar decisiones, realizar inferencias y sacar conclusiones válidas. Una falta de confiabilidad puede introducir errores y sesgos en los análisis y afectar la validez de los resultados.

*¿Qué tanto se puede repetir la medición de tal forma que produzca resultados similares en condiciones similares?*

**Validez**

Se refiere a la precisión y la exactitud con la que una medición o un conjunto de datos captura y evalúa el concepto o fenómeno que se pretende medir o evaluar. En otras palabras, la validez se refiere a si una medición realmente mide lo que pretende medir y si los resultados obtenidos son interpretados correctamente en relación con el fenómeno que se está estudiando.

La validez es un concepto esencial en la investigación y la recopilación de datos, ya que garantiza que las mediciones y los resultados sean relevantes y útiles para los propósitos de estudio. Una medición o un conjunto de datos pueden ser confiables (consistentes y reproducibles), pero aún así carecer de validez si no están midiendo de manera precisa el concepto que se desea evaluar.

*¿Se esta midiendo lo que realmente se debe medir?*

**Es necesario que los resultados de una investigación sean confiables y válidos.**

¿Qué factores afectan la confiabilidad y validez?

La confiabilidad y la validez de los datos pueden verse afectadas por una variedad de factores que pueden introducir sesgos, errores o inexactitudes en las mediciones y los resultados. Algunos de los factores más comunes que pueden influir en la confiabilidad y la validez de los datos son los siguientes:

Factores que afectan la Confiabilidad:

* Error de Medición: Incluye errores aleatorios y sistemáticos que pueden surgir durante el proceso de medición, como errores de instrumentación, errores humanos al registrar datos y variabilidad natural en las respuestas.

  **a. Error aletorio:** El error aleatorio es una variabilidad no controlada o aleatoria que puede ocurrir en las mediciones debido a factores impredecibles o fluctuaciones naturales en el sistema. Estos errores pueden ocurrir de manera aleatoria y sin un patrón predecible.

  **b. Error sistemático:** Es el error de redondeo que se lleva en cada una de las mediciones.

* Consistencia Temporal: Las mediciones pueden variar con el tiempo debido a factores como cambios en las condiciones del entorno, la salud o el estado emocional de los participantes.

* Condiciones de Medición: Las condiciones en las que se realiza la medición, como la iluminación, el ruido o la temperatura, pueden influir en la confiabilidad.

* Muestreo: La forma en que se selecciona la muestra puede afectar la confiabilidad si no es representativa de la población objetivo.

* Estabilidad de la Medición: Las mediciones pueden verse afectadas por la consistencia del instrumento de medición y la forma en que se administra.

* Error de Registro: Puede ocurrir cuando los datos se ingresan incorrectamente o se transcriben de manera errónea.

Factores que afectan la Validez:

* Validez de Contenido: La falta de representación completa de los aspectos del concepto que se está midiendo puede afectar la validez de contenido.

* Validez de Criterio: Si el criterio de comparación utilizado para evaluar la validez no es relevante o apropiado, puede afectar la validez de criterio.

* Validez de Constructo: Si el instrumento de medición no está diseñado adecuadamente para medir el constructo teórico, la validez de constructo puede verse comprometida.

* Error de Medición: Los errores de medición también pueden afectar la validez, ya que pueden llevar a mediciones inexactas de los fenómenos que se están evaluando.

* Efectos del Participante: Las respuestas de los participantes pueden verse influidas por factores como el deseo de agradar al investigador (sesgo de deseabilidad social) o las expectativas previas.

* Ambigüedad de las Preguntas: Preguntas mal redactadas o ambiguas en un cuestionario pueden llevar a respuestas incorrectas o inconsistentes.

* Validez Ecológica: Si el entorno de medición no refleja adecuadamente el contexto del mundo real, la validez ecológica puede verse comprometida.

* Sesgos Culturales o de Género: Las diferencias culturales o de género pueden influir en cómo se interpretan y responden las preguntas, afectando la validez.

Es importante reconocer estos factores y tomar medidas para minimizar su impacto en la confiabilidad y la validez de los datos. Esto implica un diseño cuidadoso de la investigación, la selección de métodos de medición apropiados, la implementación de procedimientos de muestreo rigurosos y la consideración de posibles sesgos y errores en todas las etapas del proceso de recopilación y análisis de datos.

De acuerdo a los datos, también tenemos dos características que nos permiten describir como estan representados los datos, de este modo:

1. Normalidad:  Se dice que los datos son normales si el patrón que se describe de acuerdo a la distribución de los datos describe una curva en forma de campana.
2. Anormalidad: Se dice que los datos son anormales si no tienen una representación normal.

## **Conceptos en la aplicación de los métodos estadísticos**

Para el entendimiento del presente curso, será importante tener en cuenta algunos términos importantes para la aplicación de métodos estadísticos:

1. Población: La población se refiere al conjunto completo de elementos o individuos que se están estudiando en un contexto específico. En estadística, puede ser un grupo grande y diverso, como "todos los estudiantes de una escuela" o "todos los ciudadanos de un país".

2. Muestra: Una muestra es un subconjunto seleccionado de la *población* total. Debido a que analizar toda la población puede ser impracticable o costoso, se toma una muestra representativa para hacer inferencias sobre la población en su conjunto. La calidad de la muestra y su representatividad son cruciales para obtener resultados confiables.

**Nota:** Es importante notar que la población que se puede brindar puede tener infinitos elementos o también ser finitos, mientras que las muestras siempre considera una cantidad finita. De igual modo es relevante tener en cuenta la población base donde son tomados los datos, para el estudio adecuado de la misma.

3. Parámetro: Un parámetro es una medida numérica que describe una característica específica de una *población*. Por ejemplo, la media (promedio) de edades de todos los estudiantes en una escuela sería un parámetro de la población total.

4. Estadístico o Estadígrafo: Una medida numérica calculada a partir de una muestra, que se utiliza para describir una característica particular de esa *muestra*. Ejemplos de estadísticos son la media muestral, la mediana muestral, la varianza muestral, entre otros.

**Nota:** La principal diferencia entre un parámetro y un estadístico es que un parámetro es una medida que describe una característica de la población completa y es desconocida, mientras que un estadístico es una medida calculada a partir de una muestra y se utiliza para hacer estimaciones sobre los parámetros poblacionales. Los estadísticos son herramientas fundamentales en la estadística inferencial, que busca obtener información sobre una población a partir de una muestra.

5. Estimación: La estimación es el proceso de inferir o calcular un valor aproximado de un parámetro poblacional a partir de datos de muestra. Las estimaciones se basan en la información recopilada en la muestra y se utilizan para hacer inferencias sobre la población en su conjunto.

6. Distribución de probabilidad: Una distribución de probabilidad describe cómo se distribuyen los valores posibles de una variable aleatoria (i.e. es  la forma de agrupación de los datos). En estadística, estas distribuciones se utilizan para modelar y entender el comportamiento de los datos. Algunos ejemplos de distribuciones de probabilidad son la distribución normal, la distribución binomial, la distribución de Poisson, la distribución geométrica, distribución hipergeométrica, entre otras (veasé una lista de algunas distribuciones en https://keepcoding.io/blog/tipos-distribuciones-estadistica/).


## **Estadística descriptiva vs Estadística inferencial**

Los métodos estadísticos se pueden clasificar en dos fases principales:

1. Estadística Descriptiva.
2. Estadística Inferencial.

No obstante, esta clasificación no implica la existencia de dos tipos separados de estadística. Más bien, la estadística descriptiva desempeña un papel fundamental al proporcionar una guía y un camino hacia la estadística inferencial. Esta última se basa en las técnicas y resultados de la primera, lo que nos permite generalizar conclusiones de una muestra a una población más amplia, fundamentados en la teoría de la probabilidad.

### **1. La estadística descriptiva** se centra en resumir y describir las características fundamentales de un conjunto de datos. Su objetivo principal es organizar, resumir y presentar los datos de manera que se puedan entender fácilmente.

En este aspecto se revisan:
* Medidas de tendencia central.
* Medidas de dispersión.
* Gráficos y visualizaciones.

### **2. La estadística inferencial** se utiliza para hacer inferencias o deducciones sobre una población más amplia basándose en muestras de datos. En lugar de simplemente describir los datos, la estadística inferencial busca hacer predicciones, tomar decisiones y sacar conclusiones sobre una población en función de la información de una muestra.

En este aspecto se revisan:
* Muestreo.
* Estimación.
* Pruebas de hipótesis.
* Intervalos de confianza.

### **Cualitativas**

Las variables cualitativas, también conocidas como variables categóricas o variables nominales, son tipos de variables en estadísticas y análisis de datos que representan categorías o etiquetas en lugar de valores numéricos. Estas variables se utilizan para describir características que no pueden cuantificarse directamente en términos numéricos.

* **Variables Cualitativas Nominales:** Estas variables representan categorías sin ningún tipo de orden inherente. Ejemplos de variables nominales incluyen el género (masculino, femenino), el color (rojo, azul, verde), la nacionalidad (estadounidense, mexicano, japonés) o el tipo de animal (perro, gato, pájaro).

* **Variables Cualitativas Ordinales:** Estas variables también representan categorías, pero en este caso, tienen un orden o jerarquía específica entre las categorías. Sin embargo, la distancia entre las categorías no es necesariamente uniforme ni cuantificable. Ejemplos de variables ordinales son el nivel educativo (primaria, secundaria, universidad) o la calificación de satisfacción (muy insatisfecho, insatisfecho, neutral, satisfecho, muy satisfecho).

Estas variables son importantes en la investigación y el análisis de datos porque permiten agrupar y clasificar datos en categorías significativas. Sin embargo, debido a que no se pueden realizar operaciones numéricas directas con ellas, se utilizan principalmente para propósitos descriptivos y de análisis exploratorio, como tabulaciones cruzadas, gráficos de barras, gráficos de torta y pruebas estadísticas específicas para datos categóricos.

Es crucial comprender la naturaleza de las variables cualitativas y elegir las herramientas de análisis adecuadas según el tipo de variable y los objetivos del estudio.

### **Cuantitativas**

Las variables cuantitativas, también conocidas como variables numéricas o variables ordinales, son tipos de variables en estadísticas y análisis de datos que representan valores numéricos medibles. Estas variables permiten realizar operaciones matemáticas y aritméticas, lo que las diferencia de las variables cualitativas. Las variables cuantitativas se dividen en dos tipos principales: discretas y continuas.

* **Variables Cuantitativas Discretas:** Estas variables toman valores numéricos específicos y separados, generalmente en forma de números enteros. No pueden tomar valores intermedios entre dos puntos específicos. Ejemplos de variables cuantitativas discretas son la cantidad de personas en una familia, el número de estudiantes en una clase o el número de autos en un estacionamiento.

* **Variables Cuantitativas Continuas:** Estas variables pueden tomar cualquier valor dentro de un rango continuo. No hay límites claros y separados entre los valores posibles. Estas variables a menudo involucran medidas de magnitudes, como longitud, peso, tiempo y temperatura. Ejemplos de variables cuantitativas continuas son la altura de las personas, el peso de un objeto, la temperatura en grados Celsius, etc.

Las variables cuantitativas son fundamentales en la investigación y el análisis de datos, ya que permiten aplicar técnicas estadísticas más avanzadas, como cálculos de promedio, mediana, desviación estándar, correlación, regresión y pruebas de hipótesis. Estas variables proporcionan una mayor cantidad de información numérica y, por lo tanto, son ampliamente utilizadas para tomar decisiones basadas en datos y realizar análisis más detallados de patrones y tendencias.

Es esencial comprender la distinción entre variables cualitativas y cuantitativas, ya que influye en la elección de las técnicas de análisis apropiadas y en la interpretación de los resultados obtenidos.

## **Métodos paramétricos y no paramétricos**

Los métodos paramétricos y no paramétricos son dos enfoques distintos en estadísticas y análisis de datos que se utilizan para analizar y tomar decisiones basadas en información recopilada de muestras o poblaciones. La diferencia principal entre ellos radica en cómo hacen suposiciones sobre la distribución subyacente de los datos.

* **Métodos Paramétricos:**

Los métodos paramétricos se basan en suposiciones específicas sobre la distribución de los datos, generalmente asumiendo que siguen una distribución estadística conocida, como la distribución normal. Estas suposiciones permiten que los métodos paramétricos realicen inferencias y estimaciones precisas. Algunos ejemplos de métodos paramétricos incluyen la prueba t de Student, el análisis de varianza (ANOVA), la regresión lineal y la regresión logística.

*Ventajas de los métodos paramétricos:*

Pueden proporcionar estimaciones precisas en situaciones donde las suposiciones subyacentes son válidas.
Requieren tamaños de muestra relativamente pequeños para obtener resultados significativos y precisos.

*Limitaciones de los métodos paramétricos:*

Los resultados pueden ser sesgados si las suposiciones subyacentes no se cumplen.
No son adecuados para datos que no siguen la distribución asumida.

* **Métodos No Paramétricos:**

Los métodos no paramétricos, por otro lado, no hacen suposiciones explícitas sobre la distribución de los datos subyacentes. Se utilizan cuando los datos no cumplen con las suposiciones de normalidad u otras distribuciones conocidas. Estos métodos se basan en rangos, órdenes y relaciones de orden de los datos en lugar de sus valores numéricos exactos. Ejemplos de métodos no paramétricos son la prueba de Mann-Whitney U, la prueba de Wilcoxon, la prueba de Kruskal-Wallis y la prueba de Friedman.

*Ventajas de los métodos no paramétricos:*

Son más flexibles y robustos frente a violaciones de suposiciones distribucionales.
Pueden aplicarse a una variedad más amplia de tipos de datos.

*Limitaciones de los métodos no paramétricos:*

Pueden tener menos poder estadístico en comparación con los métodos paramétricos cuando las suposiciones son válidas.
Los resultados pueden ser menos precisos cuando las muestras son pequeñas.
En resumen, la elección entre métodos paramétricos y no paramétricos depende de las características de los datos, las suposiciones subyacentes y los objetivos del análisis. En situaciones donde las suposiciones paramétricas son razonables y se cumplen, los métodos paramétricos pueden ofrecer resultados más precisos y poderosos. Sin embargo, cuando las suposiciones no se cumplen, los métodos no paramétricos pueden ser más apropiados y confiables.

## **Construcción de un modelo estadístico**

Se pueden plantear como una guía las siguientes etapas de procesos:
1. Caracterización del problema: en este punto de debe definir el problema que se va a resolver, hacer un planteamiento de las ideas desde lo más simple hasta lo más complejo.
2. Definición de objetivos e hipotesis: Se deben plantear los objetivos e hipótesis generales relacionados con el problema que se esta investigando. En este punto se define posibles caminos como se debe atacar el problema.
3. Marco teórico: Se deben exponer los elementos teóricos con los cuales se va a llevar a cabo la investigación.
4. Diseño de metodologías estadísticas: Se caracterizan las herramientas estadísticas a implementar, se observan las multiples variables, parámetros y se plantea la importancia de los mismos para realizar un proceso estructurado para generar conclusiones, observar errores y descartar procesos.
